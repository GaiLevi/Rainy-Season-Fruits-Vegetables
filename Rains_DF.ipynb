```python
import json
import pandas as pd
```

```python

with open('ims-results-json.json') as f:
    data  = json.load(f)
```

```python
df = pd.DataFrame(data)
df.drop('rain_days_code', inplace=True, axis=1)
df.drop('rain_cmnt_code', inplace=True, axis=1)
df.drop('stn_num', inplace=True, axis=1)
df['time_obs']=df['time_obs'].str.replace("T00:00:00", "")
for i in range(1,10):
    df['time_obs']=df['time_obs'].str.replace("-0"+str(i)+"-01", "-0"+str(i))
for i in range(0,3):
    df['time_obs']=df['time_obs'].str.replace("-1"+str(i)+"-01", "-1"+str(i))
df
```

```python

#df['rain_cmnt_code'].value_counts()
```

```python
df[["year","month"]] = df["time_obs"].str.split("-", expand = True)
del df['time_obs']
for i in range(1,10):
    df['month']=df['month'].str.replace("0"+str(i), str(i))
df
```
